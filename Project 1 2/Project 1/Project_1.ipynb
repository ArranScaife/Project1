The fact that consumer does not purchase anything can be 
interpreted as that she chose an outside option. Given that the fact that she chose an outside option is not recorded in this dataset, argue how you would construct a proxi variable for the choice of an outside option. Add such a proxi variable to your dataframe. 

**Hint:** you can use information that some consumers do not appear in the data every week. 

**Please input your answer in this cell:*

We have decided to implement a dictionary of dictionaries in order to help create our proxi variable. 

The inner dictionary is a dictionary of the customer IDs and it stores the values of the number of times we have seen this customer_ID purchase the item (at the end, we can see how many times a customer has bought the SKU). The outside dictionary is a dictionary of SKUs. 
Thus, a person would first call the outside dictionary looking for a specific SKU, and then call a specific Customer_ID to find the total number of times this specific customer_id has purchased this SKU. That number is our proxi variable.

Likewise, we can use this proxi variable as representation of our utility. Since it is nearly impossible to measure utility, we can use the number of times this item is bought as our utility measure. We can claim, for example, that if a given consumer bought this item 17 times, we can claim the item gives the user 17 utils vs 0 utils if the item was never bought by this consumer. Though this is rather imperfect, we find this to best estimator for utility given our limited data. For example, we argue that a customer has purchased an item 17 times because that item gives her more utility (as opposed to the same customer purchasing an item 5 times). If a customer has never purchased a specific SKU, that number if 0, indicating an outside option is taken. We can also interpret this as that item providing 0 utility to that specific customer (hence why it was never bought by that consumer).
